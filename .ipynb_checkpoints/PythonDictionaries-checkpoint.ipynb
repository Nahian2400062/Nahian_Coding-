```python
# Dictionaries
import datetime
import matplotlib.pyplot as plt
```

```python
# Sample data
data = [
    {"Date": "1/1/1989", "IMP": 37513.2},
    {"Date": "2/1/1989", "IMP": 38561.7},
    {"Date": "3/1/1989", "IMP": 39724.7},
    {"Date": "4/1/1989", "IMP": 38664.7},
    {"Date": "5/1/1989", "IMP": 40909.6},
    {"Date": "12/1/2018", "IMP": 215790.541},
    {"Date": "1/1/2019", "IMP": 209471.5167},
    {"Date": "2/1/2019", "IMP": 208905.9245},
    {"Date": "3/1/2019", "IMP": 212368.2926},
    {"Date": "4/1/2019", "IMP": 206987.8981},
]
```

```python
# Converting date strings to datetime objects
for entry in data:
    entry["Date"] = datetime.datetime.strptime(entry["Date"], "%m/%d/%Y")
```

```python
# Filter data between 1/1/2018 and 12/31/2018
start_date = datetime.datetime.strptime("1/1/2018", "%m/%d/%Y")
end_date = datetime.datetime.strptime("12/31/2018", "%m/%d/%Y")
filtered_data = [entry for entry in data if start_date <= entry["Date"] <= end_date]
```

```python
# Calculate average IMP
total_imp = sum(entry["IMP"] for entry in data)
average_imp = total_imp / len(data)
```

```python
# Plotting the data
dates = [entry["Date"] for entry in data]
imp_values = [entry["IMP"] for entry in data]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(dates, imp_values, marker="o", linestyle="-")
plt.title("IMP over Time")
plt.xlabel("Date")
plt.ylabel("IMP")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Output
print(f"Average IMP: {average_imp:.2f}")
```
